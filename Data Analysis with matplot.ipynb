2. Import Required Libraries


Start by importing Matplotlib and any other necessary libraries like Pandas and NumPy.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

3. Basic Plotting

You can create basic plots such as line plots, bar charts, scatter plots, histograms, and more using Matplotlib.

Example 1: Line Plot

```python
# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.plot(x, y)
plt.title('Line Plot of Sine Function')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

Example 2: Scatter Plot

```python
# Create scatter plot data
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot
plt.scatter(x, y, color='red')
plt.title('Scatter Plot Example')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

Example 3: Bar Plot

```python
# Bar chart data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

# Create a bar plot
plt.bar(categories, values)
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

Example 4: Histogram

```python
# Generate random data
data = np.random.randn(1000)

# Create a histogram
plt.hist(data, bins=30)
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

4. Plotting Data from Pandas DataFrame

You can also visualize data from a Pandas DataFrame.

Example: Line Plot using DataFrame

```python
# Load a sample dataset
df = pd.DataFrame({
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sales': [200, 250, 300, 350, 400]
})

# Plotting the data from DataFrame
plt.plot(df['Year'], df['Sales'], marker='o')
plt.title('Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

5. Customize Plot Appearance

Matplotlib offers many options to customize the appearance of plots such as adding labels, changing colors, adding legends, and more.

Example: Customize Plot with Labels, Grid, and Legend

```python
# Create data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot with labels, legend, and grid
plt.plot(x, y1, label='Sine', color='blue')
plt.plot(x, y2, label='Cosine', color='green')
plt.title('Sine and Cosine Waves')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()  # Add legend
plt.grid(True)  # Add grid
plt.show()
```

6. Subplots

You can create multiple plots (subplots) in a single figure using plt.subplots().

Example: Subplots

```python
# Create a 1x2 subplot
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Plot data in each subplot
x = np.linspace(0, 10, 100)
ax[0].plot(x, np.sin(x), color='blue')
ax[0].set_title('Sine Plot')

ax[1].plot(x, np.cos(x), color='green')
ax[1].set_title('Cosine Plot')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

7. Advanced Visualization Techniques

Matplotlib allows you to create more advanced visualizations like box plots, heatmaps, and 3D plots.

Example 1: Box Plot

```python
# Generate random data
data = np.random.randn(100)

# Create a box plot
plt.boxplot(data)
plt.title('Box Plot Example')
plt.show()
```

Example 2: Heatmap (with imshow)

```python
# Create data for heatmap
matrix = np.random.rand(10, 10)

# Create a heatmap
plt.imshow(matrix, cmap='hot', interpolation='nearest')
plt.colorbar()  # Add color bar
plt.title('Heatmap Example')
plt.show()
```

Example 3: 3D Plot (with mpl_toolkits)

```python
from mpl_toolkits.mplot3d import Axes3D

# Generate 3D data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title('3D Plot Example')
plt.show()
```

8. Save the Plot

You can save your plot as an image file (e.g., PNG, JPG) using the savefig() function.

```python
# Save the current plot as an image file
plt.plot(x, y)
plt.title('Save Plot Example')
plt.savefig('plot_example.png')  # Save as PNG
```
