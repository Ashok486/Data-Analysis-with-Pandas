```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Employees Table.csv')

# Display the first few rows to get a glimpse of the data
df.head()
```

```python
df.isnull().sum()
```

```python
# Fill missing values with 0
df.fillna(0, inplace=True)
```

```python
# Grouping by last name and summing salary
SALARY_by_LAST_NAME = df.groupby('LAST_NAME')['SALARY'].sum()

# Display the total sales by product
print(SALARY_by_LAST_NAME)
```

```python
# Calculate SALARY per JOB_ID
SALARY_by_JOB_ID = df.groupby('JOB_ID')['SALARY'].sum()
print(SALARY_by_JOB_ID)
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# Plotting SALARY by JOB_ID
SALARY_by_JOB_ID.plot(kind='bar', color='green')

# Adding labels and title
plt.title('Total SALARY by JOB_ID')
plt.xlabel('JOB_ID')
plt.ylabel('SALARY')
plt.show()
```

```python
import seaborn as sns
```

```python
import seaborn as sns

# Calculate the correlation matrix
corr_matrix = df[['MANAGER_ID', 'DEPARTMENT_ID']].corr()

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Between MANAGER_ID and DEPARTMENT_ID')
plt.show()
```

```python
# Group saLARY by HIRE_date
SALARY_by_HIRE_DATE = df.groupby('JOB_ID')['SALARY'].sum()

# Plot sales trend over time
SALARY_by_HIRE_DATE.plot(figsize=(10, 6))
plt.title('Sales Trend Over Time')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```
